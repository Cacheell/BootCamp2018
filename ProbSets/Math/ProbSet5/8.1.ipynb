```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-5, 30, 100)
y = np.linspace(-5, 25, 100)
X, Y = np.meshgrid(x, y)

z = 5*X - 4*Y

plt.plot(x, (2*x + 4)/3, 'r')
plt.plot(x, (x - 1) / 6, 'r')
plt.plot(x, 6 - x, 'r')
plt.plot(x, np.zeros_like(x), 'r')
plt.plot(np.zeros_like(y), y, 'r')
plt.contour(X, Y, z, 50, cmap = "binary")
plt.colorbar()

plt.xlim(-1, 8)
plt.ylim(-1, 8)
#plt.axis("equal")

plt.show()
```

Therefore the optimizer for this problem is at the point around (5, 1)

 ## Exercise 8.2.ii)

```python
x = np.linspace(-10, 50, 100)
y = np.linspace(-10, 50, 100)
X, Y = np.meshgrid(x, y)

z = 4*X + 6*Y

plt.plot(x, x + 11 , 'b')
plt.plot(x, 27 - x, 'y')
plt.plot(x, - (2*x - 90) / 5, 'm')
plt.plot(x, np.zeros_like(x), 'r')
plt.plot(np.zeros_like(y), y, 'r')
plt.contour(X, Y, z, 25, cmap = "binary")
plt.colorbar()

plt.xlim(-1, 30)
plt.ylim(-1, 30)
#plt.axis("equal")

blue_mag = np.array([[-1, 1], [2, 5]])
blue_mag_b = np.array([11, 90])
x = np.linalg.solve(blue_mag, blue_mag_b)
print("Vertix blue line and magenta:", 4*x[0] + 6*x[1])

yellow_mag = np.array([[1, 1], [2, 5]])
yellow_mag_b = np.array([27, 90])
x = np.linalg.solve(yellow_mag, yellow_mag_b)
print("Vertix Yellow line and magenta", 4*x[0] + 6*x[1])
print(x)
yellow_red = np.array([[1, 1], [0, 1]])
yellow_red_b = np.array([27, 0])
x = np.linalg.solve(yellow_red, yellow_red_b)
print("Vertix Yellow line and Red", 4*x[0] + 6*x[1])

blue_red = np.array([[-1, 1], [1, 0]])
blue_red_b = np.array([11, 0])
x = np.linalg.solve(yellow_red, yellow_red_b)
print("Vertix Blue line and Red", 4*x[0] + 6*x[1])


plt.show()
```

The polygon is under the blue, yellow and magenta lines and above and to the right of the red axis lines. So the optimal point is  (15, 12)
